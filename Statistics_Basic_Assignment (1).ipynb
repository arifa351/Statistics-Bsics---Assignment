# Theory Questions

1. What is statistics, and why is it important?
 - Statistics is the science of collecting, organizing, analyzing, and
  interpreting data. It's essential for decision-making in fields like business, science, healthcare, and government.

2. What are the two main types of statistics?
 - Descriptive Statistics - Summarizes and describes data.
 - Inferential Statistics - Makes predictions or inferences about a population
  based on a sample.

3. What are descriptive statistics?
 - These are methods for summarizing data, including measures like mean,  
   median, mode, range, variance, and standard deviation.

4. What is inferential statistics?
 - It uses sample data to draw conclusions or make inferences about a larger
   population, often involving probability.

5. What is sampling in statistics?
 - Sampling is the process of selecting a subset of individuals from a     population to estimate characteristics of the whole group.

6. What are the different types of sampling methods?
 - Random Sampling
 - Stratified Sampling
 - Systematic Sampling
 - Cluster Sampling
 - Convenience Sampling

7. What is the difference between random and non-random sampling?
 - Random Sampling: Every member has an equal chance of selection.
 - Non-Random Sampling: Selection is based on criteria or convenience, not probability.

8. Define and give examples of qualitative and quantitative data.
 - Qualitative (Categorical): Describes qualities — e.g., colors, brands.
 - Quantitative (Numerical): Measures quantities — e.g., age, weight, height.

9. What are the different types of data in statistics?
 - Qualitative: Nominal, Ordinal
 - Quantitative: Discrete, Continuous

10. Explain nominal, ordinal, interval, and ratio levels of measurement.
 - Nominal: Categories only (e.g., gender).
 - Ordinal: Categories with order (e.g., rankings).
 - Interval: Numeric, no true zero (e.g., temperature).
 - Ratio: Numeric with true zero (e.g., weight, height).

11. What is the measure of central tendency?
 - It describes the center of a data set (Mean, Median, Mode).

12. Define mean, median, and mode.
 - Mean: Average
 - Median: Middle value
 - Mode: Most frequent value

13. What is the significance of the measure of central tendency?
 - It gives a single representative value that summarizes the entire dataset.

14. What is variance, and how is it calculated?
 - Variance measures how spread out data points are from the mean.
 - Formula: σ² = Σ(x - μ)² / N (population)

15. What is standard deviation, and why is it important?
 - Standard deviation is the square root of variance. It indicates how much data deviates from the mean.

16. Define and explain the term range in statistics.
 - Range is the difference between the highest and lowest values in a dataset.
 - Formula: Range = Max - Min

17. What is the difference between variance and standard deviation?
 - Variance: Average squared deviations.
 - Standard Deviation: Square root of variance, in same units as data.

18. What is skewness in a dataset?
 - Skewness measures asymmetry in data distribution.

19. What does it mean if a dataset is positively or negatively skewed?
 - Positively Skewed: Tail on right; mean > median
 - Negatively Skewed: Tail on left; mean < median

20. Define and explain kurtosis.
 - Kurtosis measures the "tailedness" of a distribution.
 - High kurtosis: Heavy tails (outliers)
 - Low kurtosis: Light tails

21. What is the purpose of covariance?
 - Covariance indicates the direction of a linear relationship between two variables.

22. What does correlation measure in statistics?
 - Correlation measures both the strength and direction of a linear relationship. Values range from -1 to 1.

23. What is the difference between covariance and correlation?
 - Covariance: Unscaled; shows direction
 - Correlation: Scaled; shows direction and strength

24. What are some real-world applications of statistics?
 - Healthcare: Drug efficacy analysis
 - Business: Market analysis, forecasting
 - Government: Census, policy planning
 - Education: Student performance metrics
 - Sports: Performance analytics

```python
# Solution 1

import statistics

data = [12, 15, 12, 15, 14, 16, 16, 16, 18]

mean = statistics.mean(data)
median = statistics.median(data)
mode = statistics.mode(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
# Solution 2

import statistics

data = [12, 15, 14, 16, 18, 20]

variance = statistics.variance(data)
std_dev = statistics.stdev(data)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# Solution 4

import pandas as pd
from sklearn.model_selection import train_test_split

# Create sample DataFrame
df = pd.DataFrame({
    'Name': ['A','B','C','D','E','F','G','H'],
    'Gender': ['M','F','M','F','F','M','F','M'],
    'Score': [45, 55, 65, 75, 85, 95, 55, 65]
})

# Random Sampling
random_sample = df.sample(n=4)

# Stratified Sampling by Gender
stratified_sample = df.groupby('Gender', group_keys=False).apply(lambda x: x.sample(2))

print("Random Sample:\n", random_sample)
print("\nStratified Sample:\n", stratified_sample)
```

```python
# Solution 7

from scipy.stats import skew, kurtosis
import numpy as np

data = np.random.normal(0, 1, 1000)
print("Skewness:", skew(data))
print("Kurtosis:", kurtosis(data))
```

```python
# Solution 8

import matplotlib.pyplot as plt
import seaborn as sns

pos_skew = np.random.exponential(scale=2, size=1000)
neg_skew = -np.random.exponential(scale=2, size=1000)

sns.histplot(pos_skew, kde=True)
plt.title("Positive Skew")
plt.show()

sns.histplot(neg_skew, kde=True, color='red')
plt.title("Negative Skew")
plt.show()
```

```python
# Solution 9

x = [2, 4, 6, 8]
y = [1, 3, 2, 5]

cov_matrix = np.cov(x, y)
print("Covariance:\n", cov_matrix)
```

```python
# Solution 10

from scipy.stats import pearsonr, spearmanr

pearson_corr, _ = pearsonr(x, y)
spearman_corr, _ = spearmanr(x, y)

print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```

```python
# Solution 13

import statistics as stats

classes = ['0-10', '10-20', '20-30', '30-40']
frequencies = [5, 10, 15, 20]
midpoints = [5, 15, 25, 35]

# Mean (Grouped)
grouped_mean = sum([f * m for f, m in zip(frequencies, midpoints)]) / sum(frequencies)
print("Grouped Mean:", grouped_mean)
```
